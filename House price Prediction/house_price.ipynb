```python
df1=df.drop(['yr_renovated', 'street', 'city',
       'statezip', 'country','date'], axis=1)
```

```python
df1.info()
```

```python
#check outliers

import scipy.stats as stats
z_scores = stats.zscore(df1)
threshold = 3
print("size of original data:", df1.shape)
outliers = (z_scores > threshold).any(axis=1)
df1 = df1[~outliers]
print("size of data after removing outliers:", df1.shape)
```

```python
df1.head()
```

```python
from sklearn.preprocessing import StandardScaler #convert data into standard format
scaler = StandardScaler()
scaler.fit(df1)
df1_scaled = pd.DataFrame(scaler.transform(df1), columns=df1.columns)

```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python

X = df1_scaled.drop('price', axis=1)
y = df1_scaled['price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
print("X_train shape:", X_train.shape)
print("X_test shape:", X_test.shape)
print("y_train shape:", y_train.shape)
print("y_test shape:", y_test.shape)
```

```python
models = [("Linear Regression", LinearRegression()),
          ("KNN Regressor", KNeighborsRegressor()),
          ("Decision Tree Regressor", DecisionTreeRegressor()),
          ("Random Forest Regressor", RandomForestRegressor())]

```

```python
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error,root_mean_squared_error
```

```python
for name, model in models:
    print(f"Training {name}...")
    print("------------------------------")
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    print(f"{name} R^2 Score: {r2_score(y_test, y_pred):.4f}")
    print(f"{name} MAE: {mean_absolute_error(y_test, y_pred):.4f}")
    print(f"{name} RMSE: {root_mean_squared_error(y_test, y_pred):.4f}")
    print(f"{name} MSE: {mean_squared_error(y_test, y_pred):.4f}")
    print("\n")
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=[name for name, _ in models], y=[r2_score(y_test, model.predict(X_test)) for _, model in models])
plt.title('R^2 Score Comparison of Models')
plt.xlabel('Model')
plt.ylabel('R^2 Score')
plt.ylim(0, 1)
plt.show()
```
