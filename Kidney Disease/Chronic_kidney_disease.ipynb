```python
#convert object to float data type
text_columns = ['packed_cell_volume',	'white_blood_cell_count',	'red_blood_cell_count']
for i in text_columns:
    print(f"{i} : {df[i].dtype}")
```

```python
def convert_text_to_numeric(df, column ):
    df[column] = pd.to_numeric(df[column], errors='coerce')

for column in text_columns:
    convert_text_to_numeric(df, column)
    print(f"{column} : {df[column].dtype}")
```

```python
#handle missing values
missing = df.isnull().sum()
missing[missing > 0].sort_values(ascending=False)
```

```python
#find mean value
def mean_value(df, column):
    mean = df[column].mean()
    df[column].fillna(mean, inplace=True)
def mode_value(df, column):
    mode = df[column].mode()[0]
    df[column].fillna(mode, inplace=True)
def median_value(df, column):
    median = df[column].median()
    df[column].fillna(median, inplace=True)
for column in missing[missing > 0].index:
    if df[column].dtype == 'float64' or df[column].dtype == 'int64': #numerical data
        mean_value(df, column)
    elif df[column].dtype == 'object':  #categorical data
        mode_value(df, column)
    else:  #other data types
        median_value(df, column)
```

```python
df['diabetes_mellitus'] = df['diabetes_mellitus'].replace({' yes': 'yes', '\tno': 'no' , '\tyes': 'yes'})
df['coronary_artery_disease'] = df['coronary_artery_disease'].replace({' no': 'no', '\tno': 'no' , '\tyes': 'yes'})
df['class'] = df['class'].replace({'ckd\t': 'ckd', 'notckd': 'notckd'})
print(f'diabetes_mellitus : {df["diabetes_mellitus"].unique()}')
print(f'coronary_artery_disease : {df["coronary_artery_disease"].unique()}')
print(f'class : {df["class"].unique()}')
```

```python
#encoding categorical data using mapping method
df['diabetes_mellitus'] = df['diabetes_mellitus'].map({'yes': 1, 'no': 0})
df['coronary_artery_disease'] = df['coronary_artery_disease'].map({'yes': 1, 'no': 0})
df['class'] = df['class'].map({'ckd': 1, 'notckd': 0})
df['red_blood_cells'] = df['red_blood_cells'].map({'normal': 1, 'abnormal': 0})
df['pus_cell'] = df['pus_cell'].map({'normal': 1, 'abnormal': 0})
df['pus_cell_clumps'] = df['pus_cell_clumps'].map({'notpresent': 0, 'present': 1})
df['bacteria'] = df['bacteria'].map({'notpresent': 0, 'present': 1})
df['appetite'] = df['appetite'].map({'good': 1, 'poor': 0})
df['peda_edema'] = df['peda_edema'].map({'no': 0, 'yes': 1})
df['anemia'] = df['anemia'].map({'no': 0, 'yes': 1})
df['hypertension'] = df['hypertension'].map({'yes': 1, 'no': 0})
df.head(5)
```

```python
x=df.drop('class', axis=1)
y=df['class']

print(x.shape)
print(y.shape)
```

```python
from sklearn.model_selection import train_test_split
x_test,x_train,y_test,y_train = train_test_split(x,y, test_size=0.25, random_state=42)
print(x_test.shape)
print(x_train.shape)
print(y_test.shape)
print(y_train.shape)
```

```python


import sklearn.tree
model = sklearn.tree.DecisionTreeClassifier()
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
#compare y_test and y_pred
print(y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
```

```python
models=[]
models.append(('Naive Bayes', GaussianNB()))
models.append(('K-Nearest Neighbors', KNeighborsClassifier()))
models.append(('Support Vector Machine', SVC()))
models.append(('Random Forest', RandomForestClassifier()))
models.append(('Decision Tree', sklearn.tree.DecisionTreeClassifier()))

```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix,precision_score, recall_score, f1_score
```

```python
for name,model in models: 
    print(name,model)
    print('-----------------------------')  
    model.fit(x_train,y_train)
    y_pred=model.predict(x_test)
    print(confusion_matrix(y_test,y_pred))
    print('\n')
    print(classification_report(y_test,y_pred))
    print('-----------------------------')
    print(f'{name} : accuracy: {accuracy_score(y_test,y_pred)}')
    print(f"{name} : prediction : {precision_score(y_test,y_pred)}")   
    print(f"{name} : recall : {recall_score(y_test,y_pred)}")
    print(f"{name} : f1_score : {f1_score(y_test,y_pred)}")
```

```python
plt.figure(figsize=(15,8))
plt.bar(models[0][0], accuracy_score(y_test, models[0][1].predict(x_test)))
plt.bar(models[1][0], accuracy_score(y_test, models[1][1].predict(x_test)))
plt.bar(models[2][0], accuracy_score(y_test, models[2][1].predict(x_test)))
plt.bar(models[3][0], accuracy_score(y_test, models[3][1].predict(x_test)))
plt.bar(models[4][0], accuracy_score(y_test, models[4][1].predict(x_test)))
plt.ylim(0, 1)
plt.xlabel('Models')
plt.ylabel('Accuracy')
plt.title('Model Accuracy Comparison')
plt.show()
```
